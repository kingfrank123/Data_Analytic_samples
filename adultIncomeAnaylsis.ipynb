```python
# Replace income values [ <=50k, >50k ] with 0 and 1
#a. Show how many have less than 50 k and greater than 50 k income.
#b. Use count plot to visualize it.
tmp = df5.groupby(df5['income']).count().iloc[:,-1:]
tmp
```

```python
tmp.plot(kind = 'bar')
```

```python
#c. Create a function which will return 0 and 1 according to condition.
#d. Assign returning result into newly added column named as econded income
#e. Do the same with using replace method.
def incomeChange(x):  
    if x['income'] == '<=50K':
      return 0
    else:
      return 1
df5['econded income'] = df5.apply(lambda row: incomeChange(row), axis=1)
```

```python
df5['econded income using replace'] = df5['income'].copy()
df5.head()
```

```python
#df5['econded income using replace'] = df5['income'].replace(['<=50K', '>50K	'], [0,1])
```

```python
df5['econded income using replace'] = df5['econded income using replace'].replace(['<=50K'], '0')
df5['econded income using replace'] = df5['econded income using replace'].replace(['>50K'], '1')
df5.head()
```

```python
#3. Which workclass getting the highest income?
#4. How has better chance to get income greater than 50k male or female?
#5. Convert workclass columns datatype to category datatype
#a. Converting datatype can be useful in case of saving memory, you
#need to optimize it.
# grouping by workclass then counting the econded income column
workclasses = df5.groupby([df5["workclass"],df5['econded income']])
workclasses.sum().iloc[:,-1:]
# looks like the highest income goes to the Private work class
```

```python
# 4. How has better chance to get income greater than 50k male or female?
gender = df5.groupby([df5["gender"],df5['econded income']])
gender.count().iloc[:,-1:]
# looks like male has higher income chance at 1(>50k)
```
